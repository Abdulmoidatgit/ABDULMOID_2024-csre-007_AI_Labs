```python
import numpy as np

marks = np.array([35, 50, 72, 28, 65, 90])
labels = []

for m in marks:
    if m >= 40:
        labels.append("Pass")
    else:
        labels.append("Fail")

print("Marks:", marks)
print("Labels:", labels)
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

marks = np.array([40, 45, 48, 55, 60, 70, 80, 90])

new_marks = np.array([35, 65, 75, 50, 85])
marks = np.append(marks, new_marks)

classification = np.where(marks >= 50, "Pass", "Fail")

for i, mark in enumerate(marks):
    print(f"Student {i+1}: Mark = {mark}, Classification = {classification[i]}")

plt.figure(figsize=(10,5))
plt.plot(range(1, len(marks)+1), marks, marker='o', linestyle='-', color='blue')
plt.title("Student Marks")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset: 5 students
names = ["Ali", "Sara", "Bilal", "Ayesha", "Hassan"]
marks_ai = [85, 90, 70, 95, 80]
marks_ds = [80, 85, 75, 90, 70]

plt.figure(figsize=(8,4))
plt.bar(names, marks_ai, color='skyblue')
plt.title("AI Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()

plt.figure(figsize=(8,4))
plt.bar(names, marks_ds, color='lightgreen')
plt.title("Data Structures Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()

plt.figure(figsize=(6,6))
plt.scatter(marks_ai, marks_ds, color='orange', s=100)
for i, name in enumerate(names):
    plt.text(marks_ai[i]+0.5, marks_ds[i]+0.5, name)  # label points
plt.title("AI Marks vs DS Marks")
plt.xlabel("AI Marks")
plt.ylabel("DS Marks")
plt.grid(True)
plt.show()
```
